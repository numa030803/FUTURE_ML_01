```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display plots inside notebook
%matplotlib inline





```

```python
df = pd.read_csv("../data/raw/superstore.csv", encoding="latin1")
```

```python
# Display first 5 rows
df.head()

# Shape of the dataset
df.shape

# Column names
df.columns

# Dataset information
df.info()

# Summary statistics
df.describe()

# Missing values
df.isnull().sum()
```

# Data Cleaning

The purpose of this step is to prepare the dataset for forecasting by:
- Removing duplicate records
- Converting date columns to datetime format
- Sorting data chronologically
- Verifying data quality

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Convert date columns to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df.info()
```

```python
# Sort dataset by order date
df = df.sort_values(by="Order Date")

# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
print("Earliest Order:", df["Order Date"].min())
print("Latest Order:", df["Order Date"].max())
```

This above cell gives us historical data

```python
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n")
print(df.isnull().sum())
```

# Exploratory Data Analysis (EDA)

Exploratory Data Analysis helps us understand the sales data by identifying trends, patterns, and relationships. These insights are useful for selecting appropriate features and building an effective forecasting model.

```python
# SALES DISTRIBUTION
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=30)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

The sales distribution shows how frequently different sales values occur. A right-skewed distribution indicates that most transactions involve lower sales amounts, while a few transactions have very high sales values.

```python
# SALES OVER TIME
daily_sales = df.groupby("Order Date")["Sales"].sum()
# The above line groups by Order Date, selects
# column Sales and gives us the sum

plt.figure(figsize=(14,6))
plt.plot(daily_sales)
plt.title("Daily Sales Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.show()
```

This plot helps identify long-term trends, fluctuations, and potential seasonality in sales, which are essential for forecasting future demand.

```python
# MONTHLY SALES TREND
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

monthly_sales.plot(figsize=(14,6))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Aggregating sales by month reduces daily fluctuations and makes it easier to observe overall business growth and seasonal patterns.

```python
# SALES BY CATEGORY
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

Comparing sales across product categories helps identify which categories contribute the most to overall revenue.

```python
# SALES BY REGION 
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

Regional analysis highlights geographical differences in sales performance, enabling businesses to allocate resources more effectively.

```python
# PROFIT v/s SALES
plt.figure(figsize=(8,6))
plt.scatter(df["Sales"], df["Profit"])

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

This visualization helps determine whether higher sales consistently result in higher profits and reveals transactions with high sales but low profitability.

```python
# TOP 10 PRODUCTS by SALES
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))
top_products.plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

Identifying the highest-selling products helps businesses prioritize inventory and marketing efforts.

```python
# CORRELATION HEATMAP
numeric_df = df.select_dtypes(include=["number"])

corr = numeric_df.corr()

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()
```

The correlation matrix illustrates relationships between numerical variables, helping identify features that may influence sales.
